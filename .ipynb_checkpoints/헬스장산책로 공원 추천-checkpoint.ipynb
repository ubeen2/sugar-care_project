```python
import re
import folium
import pandas as pd
import requests, os, time
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# === API KEYS (êµì²´ ê¶Œì¥) ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# ------------ ìœ í‹¸ ------------
def get_naver_image(query):
    """1ìˆœìœ„ ì´ë¯¸ì§€ ì†ŒìŠ¤: ë„¤ì´ë²„ ì´ë¯¸ì§€ API (ì‹¤íŒ¨ ì‹œ None ë°˜í™˜)"""
    url = "https://openapi.naver.com/v1/search/image"
    headers = {"X-Naver-Client-Id": NAVER_CLIENT_ID,
               "X-Naver-Client-Secret": NAVER_CLIENT_SECRET}
    params = {"query": query, "display": 1, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        res = r.json()
        if res.get("items"):
            link = res["items"][0]["link"]
            if isinstance(link, str) and link.startswith("http"):
                return link
    except:
        pass
    return None

def get_og_image_from_kakao(place_url):
    """2ìˆœìœ„(í´ë°±) ì´ë¯¸ì§€: ì¹´ì¹´ì˜¤ ì¥ì†Œ og:image"""
    try:
        h = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
        }
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']', html, re.IGNORECASE)
        if m:
            return m.group(1)
    except:
        return None
    return None

def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) ì£¼ì†Œ ê²€ìƒ‰
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) í‚¤ì›Œë“œ ê·¸ëŒ€ë¡œ
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) ìˆ«ì ì œê±° í›„ í‚¤ì›Œë“œ
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    # íƒì§€ íšŒí”¼ & UA ì§€ì •
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    """ë¬¸ìì—´ì—ì„œ ì²« ë²ˆì§¸ ì‹¤ìˆ˜í˜• ìˆ«ì ì¶”ì¶œ (ì˜ˆ: 'í‰ì  4.8/5' -> 4.8)"""
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    """ìš”ì²­: í‰ì ì€ Seleniumìœ¼ë¡œ ê°€ì ¸ì˜¤ê¸°"""
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                    )
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

def get_places(category, address, size=15, sort="accuracy", radius=3000, driver=None):
    """ì‚¬ì§„ ìš°ì„ ìˆœìœ„: ë„¤ì´ë²„ API â†’ (ì—†ìœ¼ë©´) ì¹´ì¹´ì˜¤ og:image"""
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_url = p.get("place_url")
        addr_q = p["road_address_name"] or p["address_name"]
        # 1ìˆœìœ„: ë„¤ì´ë²„ ì´ë¯¸ì§€
        img = get_naver_image(f"{p['place_name']} {addr_q}")
        # 2ìˆœìœ„: ì¹´ì¹´ì˜¤ og:image
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "ì´ë¦„": p["place_name"],
            "ì¹´í…Œê³ ë¦¬": category,
            "ì£¼ì†Œ": addr_q,
            "ìœ„ë„": float(p["y"]),
            "ê²½ë„": float(p["x"]),
            "ì „í™”ë²ˆí˜¸": p["phone"],
            "ë§í¬": place_url,
            "ëŒ€í‘œì´ë¯¸ì§€": img  # Noneì´ë©´ ì¹´ë“œ/íŒì—…ì—ì„œ ì´ë¯¸ì§€ ë¯¸ë…¸ì¶œ
        })
    df = pd.DataFrame(rows)

    # ìš”ì²­: í‰ì ì€ ì…€ë ˆë‹ˆì›€ìœ¼ë¡œ
    if sort == "rating" and driver is not None and not df.empty:
        ratings = get_kakao_ratings_bulk(driver, df["ë§í¬"].tolist())
        df["í‰ì "] = df["ë§í¬"].map(ratings)
        if df["í‰ì "].notna().sum() > 0:
            df = df.sort_values("í‰ì ", ascending=False, na_position="last")

    return df

icon_urls = {
    "í—¬ìŠ¤ì¥": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "ê³µì›": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "ë‚´ê³¼": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "ì¢…í•©ë³‘ì›": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "ë‚´ë¶„ë¹„ë‚´ê³¼": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty: 
        return "<p style='color:#777'>í‘œì‹œí•  ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.</p>"
    top5 = df.head(5)
    html = "<h3>ğŸ“Œ Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['ëŒ€í‘œì´ë¯¸ì§€']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" \
                  if pd.notna(r.get("ëŒ€í‘œì´ë¯¸ì§€")) and r.get("ëŒ€í‘œì´ë¯¸ì§€") else ""
        rating_txt = "ì •ë³´ ì—†ìŒ"
        if "í‰ì " in r and pd.notna(r["í‰ì "]):
            try:
                rating_txt = f"{float(r['í‰ì ']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['ì´ë¦„']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">ì¹´í…Œê³ ë¦¬: {r['ì¹´í…Œê³ ë¦¬']}</p>
            <p style="margin:0;font-size:12px;color:#555;">â­ í‰ì : {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['ì£¼ì†Œ']}</p>
        </div>
        """
    return html

def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000
    driver = init_driver() if sort == "rating" else None

    # ë°ì´í„° ìˆ˜ì§‘
    dfs = []
    cats = ["í—¬ìŠ¤ì¥", "ê³µì›", "ë‚´ê³¼", "ì¢…í•©ë³‘ì›", "ë‚´ë¶„ë¹„ë‚´ê³¼"] if category == "ì „ì²´" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius, driver=driver)
        if not d.empty:
            dfs.append(d)
    if driver:
        try:
            driver.quit()
        except:
            pass

    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780  # ì„œìš¸ì‹œì²­ fallback

    # ì§€ë„ ìƒì„± (iframeìœ¼ë¡œë§Œ ì¶œë ¥í•´ ì¤‘ë³µ ì§€ë„ ë°©ì§€)
    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>ë‚´ ìœ„ì¹˜</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    if dfs:
        df = pd.concat(dfs, ignore_index=True)
        for row in df.itertuples():
            icon_url = icon_urls.get(row.ì¹´í…Œê³ ë¦¬)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.ëŒ€í‘œì´ë¯¸ì§€}' width='150' style='border-radius:6px;object-fit:cover;'><br>" \
                       if getattr(row, "ëŒ€í‘œì´ë¯¸ì§€", None) else ""
            rating_txt = "ì •ë³´ ì—†ìŒ"
            if hasattr(row, "í‰ì ") and pd.notna(getattr(row, "í‰ì ")):
                try:
                    rating_txt = f"{float(getattr(row, 'í‰ì ')):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.ì´ë¦„}</b><br>"
                f"ì¹´í…Œê³ ë¦¬: {row.ì¹´í…Œê³ ë¦¬}<br>"
                f"ì£¼ì†Œ: {row.ì£¼ì†Œ}<br>"
                f"ì „í™”: {row.ì „í™”ë²ˆí˜¸}<br>"
                f"â­ í‰ì : {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.ë§í¬}' target='_blank'>ì¹´ì¹´ì˜¤ë§µì—ì„œ ë³´ê¸°</a>",
                max_width=320
            )
            folium.Marker([row.ìœ„ë„, row.ê²½ë„], icon=icon, popup=popup).add_to(m)
    else:
        df = pd.DataFrame()

    # ì§€ë„ë¥¼ íŒŒì¼ë¡œ ì €ì¥í•˜ê³  iframeìœ¼ë¡œ í‘œì‹œ
    os.makedirs("./_maps", exist_ok=True)
    map_path = f"./_maps/folium_map_{int(time.time())}.html"
    m.save(map_path)

    cards_html = top5_cards_html(df)
    full_html = f"""
    <div style="display:flex;gap:20px;align-items:flex-start;">
      <div style="flex:2;min-width:500px;">
        <iframe src="{map_path}" style="width:100%;height:600px;border:none;"></iframe>
      </div>
      <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
        {cards_html}
      </div>
    </div>
    """
    return full_html

# ------------ ìœ„ì ¯ UI ------------
address_input = Text(value="ì„œìš¸ ì¢…ë¡œêµ¬ ì„¸ì¢…ëŒ€ë¡œ 110", description="ì£¼ì†Œ:", layout={'width':'60%'})
category_dd   = Dropdown(options=["ì „ì²´","í—¬ìŠ¤ì¥","ê³µì›","ë‚´ê³¼","ì¢…í•©ë³‘ì›","ë‚´ë¶„ë¹„ë‚´ê³¼"], value="ì „ì²´", description="ë³´ê¸°:")
# ë¼ë²¨ì€ 'ê°€ê¹Œìš´ìˆœ', ê°’ì€ 'distance'
sort_dd       = Dropdown(options=[("ì •í™•ë„ìˆœ","accuracy"),("ê°€ê¹Œìš´ìˆœ","distance"),("í‰ì ìˆœ","rating")], value="accuracy", description="ì •ë ¬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="ë°˜ê²½(km):")
btn_search    = Button(description="ê²€ìƒ‰", button_style="success")
out           = Output()

def on_click(b):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
btn_search.on_click(on_click)
display(ui)
```

```python
# -*- coding: utf-8 -*-
import re, time
import folium
import pandas as pd
import requests

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Jupyter widgets
from ipywidgets import interact, Text, Dropdown

# =========================
# ğŸ”‘ API KEYS (ì±„ì›Œ ë„£ê¸°)
# =========================
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# =========================
# ğŸ”§ ìœ í‹¸
# =========================
def init_driver():
    """Selenium Chrome Headless ì´ˆê¸°í™”(íƒì§€ íšŒí”¼ ì˜µì…˜ í¬í•¨)"""
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")           # ìµœì‹  í—¤ë“œë¦¬ìŠ¤
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    # íƒì§€ íšŒí”¼ & UA ê³ ì •
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
    )
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text: str):
    """ë¬¸ìì—´ì—ì„œ ì²« ë²ˆì§¸ ì‹¤ìˆ˜í˜• ìˆ«ì ì¶”ì¶œ (ì˜ˆ: 'í‰ì  4.8/5' -> 4.8)"""
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    return float(m.group(1)) if m else None

def get_kakao_ratings_bulk(driver, place_urls):
    """
    âœ… ì¹´ì¹´ì˜¤ ì¥ì†Œ ìƒì„¸ í˜ì´ì§€ë“¤ì—ì„œ í‰ì ì„ ì¼ê´„ í¬ë¡¤ë§
    - ë™ì¼ ë“œë¼ì´ë²„ë¡œ ìˆœíšŒ â†’ ì†ë„/ì•ˆì •ì„± â†‘
    """
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            time.sleep(0.25)

            # ìµœëŒ€ 3íšŒ ì¬ì‹œë„ (ì§€ì—°ë Œë” ëŒ€ë¹„)
            for _ in range(3):
                for sel in selectors:
                    try:
                        elem = WebDriverWait(driver, 3).until(
                            EC.presence_of_element_located((By.CSS_SELECTOR, sel))
                        )
                        rating = _extract_first_float((elem.text or "").strip())
                        if rating is not None:
                            break
                    except Exception:
                        continue
                if rating is not None:
                    break
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(0.35)
        except Exception as e:
            # ë””ë²„ê¹…ìš© ë¡œê·¸(í•„ìš”ì‹œ ìœ ì§€)
            # print("í‰ì  í¬ë¡¤ë§ ì‹¤íŒ¨:", e)
            rating = None
        results[url] = rating
    return results

def get_naver_image(query):
    """
    âœ… 1ìˆœìœ„ ì´ë¯¸ì§€: ë„¤ì´ë²„ ì´ë¯¸ì§€ ê²€ìƒ‰ API
    - ê°€ì¥ ìœ ì‚¬í•œ 1ì¥ ë°˜í™˜, ì‹¤íŒ¨ ì‹œ None
    """
    url = "https://openapi.naver.com/v1/search/image"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }
    params = {"query": query, "display": 1, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        data = r.json()
        if data.get("items"):
            link = data["items"][0].get("link")
            if isinstance(link, str) and link.startswith("http"):
                return link
    except Exception:
        pass
    return None

def get_og_image_from_kakao(place_url):
    """
    2ìˆœìœ„(í´ë°±) ì´ë¯¸ì§€: ì¹´ì¹´ì˜¤ ì¥ì†Œ ìƒì„¸ í˜ì´ì§€ì˜ og:image
    """
    try:
        h = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                          "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
        }
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(
            r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']',
            html, re.IGNORECASE
        )
        if m:
            return m.group(1)
    except Exception:
        return None
    return None

# =========================
# ğŸ—º ì¥ì†Œ ê²€ìƒ‰ (ì¹´ì¹´ì˜¤ API)
# =========================
def get_places(category, query="ì¢…ë¡œêµ¬", size=5, sort="accuracy", driver=None, fetch_ratings=True):
    """
    ì¹´ì¹´ì˜¤ í‚¤ì›Œë“œ ê²€ìƒ‰
    - sort: accuracy | distance (ì¹´ì¹´ì˜¤ APIì—ì„œ ì§€ì›)
    - ì‚¬ì§„: ë„¤ì´ë²„ API ìš°ì„  â†’ ì¹´ì¹´ì˜¤ og:image í´ë°±
    - í‰ì : fetch_ratings=Trueì¼ ë•Œ Selenium ì¼ê´„ í¬ë¡¤ë§
    """
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": f"{query} {category}",
        "size": size,
        "sort": sort if sort in ["accuracy", "distance"] else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7)
        res.raise_for_status()
        data = res.json()
    except Exception as e:
        print("âŒ ì¹´ì¹´ì˜¤ API ìš”ì²­ ì‹¤íŒ¨:", e)
        return pd.DataFrame()

    docs = data.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_url = p.get("place_url")
        addr_q = p.get("road_address_name") or p.get("address_name") or ""

        # âœ… ì‚¬ì§„: ë„¤ì´ë²„ â†’ ì¹´ì¹´ì˜¤ og:image
        img = get_naver_image(f"{p.get('place_name','')} {addr_q}")
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "ì´ë¦„": p.get("place_name"),
            "ì¹´í…Œê³ ë¦¬": category,
            "ì£¼ì†Œ": addr_q,
            "ìœ„ë„": float(p.get("y")),
            "ê²½ë„": float(p.get("x")),
            "ì „í™”ë²ˆí˜¸": p.get("phone"),
            "ë§í¬": place_url,
            "ëŒ€í‘œì´ë¯¸ì§€": img
        })

    df = pd.DataFrame(rows)

    # âœ… í‰ì : ìš”ì²­ëŒ€ë¡œ Seleniumìœ¼ë¡œ (ëŒ€ëŸ‰)
    if fetch_ratings and driver is not None and not df.empty:
        ratings = get_kakao_ratings_bulk(driver, df["ë§í¬"].tolist())
        df["í‰ì "] = df["ë§í¬"].map(ratings)

    return df

# =========================
# ğŸ–¼ ë§ˆì»¤ ì•„ì´ì½˜
# =========================
icon_urls = {
    "í—¬ìŠ¤ì¥": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "ê³µì›": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "ë‚´ê³¼": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "ì¢…í•©ë³‘ì›": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "ë‚´ë¶„ë¹„ë‚´ê³¼": "https://img.icons8.com/ios/50/pills.png"
}

# =========================
# ğŸ—º ì§€ë„ ìƒì„±
# =========================
def make_map(query="ëŒ€êµ¬ ë¶êµ¬", category="ì „ì²´", sort="accuracy"):
    """
    - category: 'ì „ì²´'ë©´ 5ê°œ ì¹´í…Œê³ ë¦¬ ëª¨ë‘ ìˆ˜ì§‘ í›„ í•©ì¹¨
    - sort == 'rating'ì´ë©´ í‰ì  ê¸°ì¤€ ì •ë ¬
    """
    cats = ["í—¬ìŠ¤ì¥", "ê³µì›", "ë‚´ê³¼", "ì¢…í•©ë³‘ì›", "ë‚´ë¶„ë¹„ë‚´ê³¼"] if category == "ì „ì²´" else [category]

    # âœ… ë“œë¼ì´ë²„ë¥¼ ë‹¨ í•œ ë²ˆë§Œ ì—´ì–´ ëª¨ë“  í‰ì ì„ ì¼ê´„ ìˆ˜ì§‘
    driver = init_driver()
    try:
        dfs = []
        for cat in cats:
            # í‰ì ì€ í•­ìƒ í‘œì‹œë˜ê¸¸ ì›í•˜ë¯€ë¡œ fetch_ratings=True ìœ ì§€
            df_tmp = get_places(cat, query=query, size=5, sort=("distance" if sort=="distance" else "accuracy"),
                                driver=driver, fetch_ratings=True)
            if not df_tmp.empty:
                dfs.append(df_tmp)
    finally:
        try:
            driver.quit()
        except:
            pass

    if not dfs:
        print("âš ï¸ ì„ íƒí•œ ì¡°ê±´ì— ë§ëŠ” ì¥ì†Œê°€ ì—†ìŠµë‹ˆë‹¤.")
        return folium.Map(location=[37.5665, 126.9780], zoom_start=12)

    df = pd.concat(dfs, ignore_index=True)

    # âœ… í‰ì ìˆœ ì •ë ¬ ìš”ì²­ ì‹œ
    if sort == "rating" and "í‰ì " in df.columns:
        df = df.sort_values("í‰ì ", ascending=False, na_position="last")

    # ì§€ë„ ì¤‘ì‹¬: í‰ê·  ì¢Œí‘œ
    m = folium.Map(location=[df["ìœ„ë„"].mean(), df["ê²½ë„"].mean()], zoom_start=13)

    # ë§ˆì»¤ ìƒì„±
    for row in df.itertuples():
        if category == "ì „ì²´" or row.ì¹´í…Œê³ ë¦¬ == category:
            icon_url = icon_urls.get(row.ì¹´í…Œê³ ë¦¬)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None

            rating_txt = "ì •ë³´ ì—†ìŒ"
            if hasattr(row, "í‰ì ") and pd.notna(getattr(row, "í‰ì ")):
                try:
                    rating_txt = f"{float(getattr(row, 'í‰ì ')):.1f}"
                except Exception:
                    pass

            img_html = ""
            if getattr(row, "ëŒ€í‘œì´ë¯¸ì§€", None):
                img_html = f"<br><img src='{row.ëŒ€í‘œì´ë¯¸ì§€}' width='160' style='border-radius:6px;object-fit:cover;'>"

            popup = folium.Popup(
                f"<b>{row.ì´ë¦„}</b><br>"
                f"ì¹´í…Œê³ ë¦¬: {row.ì¹´í…Œê³ ë¦¬}<br>"
                f"ì£¼ì†Œ: {row.ì£¼ì†Œ}<br>"
                f"ì „í™”: {row.ì „í™”ë²ˆí˜¸ or '-'}<br>"
                f"â­ í‰ì : {rating_txt}"
                f"{img_html}"
                f"<br><a href='{row.ë§í¬}' target='_blank'>ì¹´ì¹´ì˜¤ë§µì—ì„œ ë³´ê¸°</a>",
                max_width=320
            )
            folium.Marker([row.ìœ„ë„, row.ê²½ë„], icon=icon, popup=popup).add_to(m)

    return m

# =========================
# ğŸ› ì¸í„°ë™í‹°ë¸Œ UI
# =========================
interact(
    make_map,
    query=Text(value="ëŒ€êµ¬ ë¶êµ¬", description="ì§€ì—­:"),
    category=Dropdown(
        options=["ì „ì²´", "í—¬ìŠ¤ì¥", "ê³µì›", "ë‚´ê³¼", "ì¢…í•©ë³‘ì›", "ë‚´ë¶„ë¹„ë‚´ê³¼"],
        value="ì „ì²´",
        description="ë³´ê¸°:"
    ),
    sort=Dropdown(
        options=[("ì •í™•ë„ìˆœ", "accuracy"), ("ê°€ê¹Œìš´ìˆœ", "distance"), ("í‰ì ìˆœ", "rating")],
        value="accuracy",
        description="ì •ë ¬:"
    )
)
```

```python
import re
import folium
import pandas as pd
import requests, os, time
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# === API KEYS (êµì²´ ê¶Œì¥) ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# ------------ ê³µí†µ ìœ í‹¸ ------------
def _normalize(s: str) -> str:
    if not isinstance(s, str):
        return ""
    return re.sub(r"[\s\-\_\|\(\)\[\]{}Â·â€¢âˆ™,.;:!?'\"<>@#%^&*+=/\\]", "", s.lower())

def extract_region_token(addr_text: str) -> str:
    """ì£¼ì†Œì—ì„œ ê²€ìƒ‰ìš© í† í° ì¶”ì¶œ: ìš°ì„ ìˆœìœ„ 'ë™' > 'êµ¬' > 'ì' > 'ë©´'"""
    if not isinstance(addr_text, str):
        return ""
    toks = [t for t in addr_text.split() if t]
    dong = [t for t in toks if t.endswith("ë™")]
    gu   = [t for t in toks if t.endswith("êµ¬")]
    eup  = [t for t in toks if t.endswith("ì")]
    myeon= [t for t in toks if t.endswith("ë©´")]
    return (dong or gu or eup or myeon or [""])[0]

# ------------ ì´ë¯¸ì§€ ìœ í‹¸ (ë„¤ì´ë²„ ì—„ê²© í•„í„°) ------------
def _is_likely_photo(link: str, w: int=None, h: int=None) -> bool:
    if not isinstance(link, str) or not link.startswith("http"):
        return False
    base = link.lower().split("?")[0]
    # í™•ì¥ì í•„í„°
    if not base.endswith((".jpg", ".jpeg", ".png", ".webp")):
        return False
    # ë„ˆë¬´ ì‘ì€ ì¸ë„¤ì¼ ì œì™¸
    try:
        if w is not None and h is not None and (w < 200 or h < 150):
            return False
    except:
        pass
    # ê¸ˆì§€ í‚¤ì›Œë“œ(ì•„ì´ì½˜/ë¡œê³ /ì§€ë„/ë°°ë„ˆ/ë©”ë‰´íŒ ë“±)
    bad = ("emoji","icon","favicon","sprite","logo","map","kakaomap","navermap","banner","menu","price","coupon","ad")
    if any(k in base for k in bad):
        return False
    return True

def _score_naver_item(item, name_norm: str, region_norm: str) -> int:
    title = item.get("title") or ""
    link  = item.get("link") or ""
    w = int(item.get("sizewidth", 0)) if item.get("sizewidth") else None
    h = int(item.get("sizeheight", 0)) if item.get("sizeheight") else None

    if not _is_likely_photo(link, w, h):
        return -999

    t_norm = _normalize(title)
    l_norm = _normalize(link)
    score = 0
    # ì´ë¦„/ì§€ì—­ í† í° ë§¤ì¹­ ê°€ì¤‘ì¹˜
    if name_norm and (name_norm in t_norm or name_norm in l_norm):
        score += 3
    if region_norm and (region_norm in t_norm or region_norm in l_norm):
        score += 2
    # í•´ìƒë„ ê°€ì (ëŒ€ëµ)
    if (w or 0) >= 400 and (h or 0) >= 300:
        score += 1
    return score

def get_naver_image_strict(place_name: str, addr_text: str):
    """
    ë„¤ì´ë²„ ì´ë¯¸ì§€ APIë§Œ ì‚¬ìš©í•˜ë˜, ìƒìœ„ 10ê°œ ì¤‘ ê°€ì¥ ê´€ë ¨ì„± ë†’ì€ ì´ë¯¸ì§€ë§Œ ì„ íƒ.
    - ì¿¼ë¦¬ëŠ” '{ë™/êµ¬} {ìƒí˜¸ëª…}' (í† í° ì—†ìœ¼ë©´ 'ìƒí˜¸ëª… ì£¼ì†Œ')
    - ì ìˆ˜ < 2 ì´ë©´ None ë°˜í™˜ â†’ ì´í›„ og:image í´ë°±
    """
    region = extract_region_token(addr_text)
    query = f"{region} {place_name}" if region else f"{place_name} {addr_text}"

    url = "https://openapi.naver.com/v1/search/image"
    headers = {"X-Naver-Client-Id": NAVER_CLIENT_ID,
               "X-Naver-Client-Secret": NAVER_CLIENT_SECRET}
    params = {"query": query, "display": 10, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        data = r.json()
        items = data.get("items", []) or []
        if not items:
            return None

        name_norm   = _normalize(place_name)
        region_norm = _normalize(region)
        best, best_score = None, -999
        for it in items:
            sc = _score_naver_item(it, name_norm, region_norm)
            if sc > best_score:
                best, best_score = it, sc

        if best and best_score >= 2:
            return best.get("link")
        return None
    except:
        return None

def get_og_image_from_kakao(place_url):
    """ì¹´ì¹´ì˜¤ ì¥ì†Œ ìƒì„¸ í˜ì´ì§€ì˜ og:imageë¥¼ í´ë°±ìœ¼ë¡œ ì‚¬ìš©."""
    try:
        h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"}
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']', html, re.IGNORECASE)
        return m.group(1) if m else None
    except:
        return None

# ------------ ì¢Œí‘œ ìœ í‹¸ ------------
def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) ì£¼ì†Œ ê²€ìƒ‰
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) í‚¤ì›Œë“œ ê·¸ëŒ€ë¡œ
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) ìˆ«ì ì œê±° í›„ í‚¤ì›Œë“œ
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

# ------------ Selenium í‰ì  í¬ë¡¤ë§ ------------
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector)))
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

# ------------ ì¥ì†Œ ê²€ìƒ‰ (ì´ë¯¸ì§€: ë„¤ì´ë²„ ì—„ê²© ì„ íƒ â†’ og:image í´ë°±) ------------
def get_places(category, address, size=15, sort="accuracy", radius=3000):
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_name = p.get("place_name")
        addr_q     = p.get("road_address_name") or p.get("address_name")
        place_url  = p.get("place_url")

        # ë„¤ì´ë²„ ì´ë¯¸ì§€(ì—„ê²©) â†’ ì‹¤íŒ¨ ì‹œ og:image
        img = get_naver_image_strict(place_name, addr_q)
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "ì´ë¦„": place_name,
            "ì¹´í…Œê³ ë¦¬": category,
            "ì£¼ì†Œ": addr_q,
            "ìœ„ë„": float(p["y"]),
            "ê²½ë„": float(p["x"]),
            "ì „í™”ë²ˆí˜¸": p.get("phone"),
            "ë§í¬": place_url,
            "ëŒ€í‘œì´ë¯¸ì§€": img
        })
    return pd.DataFrame(rows)

# ------------ ì•„ì´ì½˜ & ì¹´ë“œ ------------
icon_urls = {
    "í—¬ìŠ¤ì¥": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "ê³µì›": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "ë‚´ê³¼": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "ì¢…í•©ë³‘ì›": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "ë‚´ë¶„ë¹„ë‚´ê³¼": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty:
        return "<p style='color:#777'>í‘œì‹œí•  ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.</p>"
    top5 = df.head(5)
    html = "<h3>ğŸ“Œ Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['ëŒ€í‘œì´ë¯¸ì§€']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" if pd.notna(r.get("ëŒ€í‘œì´ë¯¸ì§€")) and r.get("ëŒ€í‘œì´ë¯¸ì§€") else ""
        rating_txt = "ì •ë³´ ì—†ìŒ"
        if "í‰ì " in r and pd.notna(r["í‰ì "]):
            try:
                rating_txt = f"{float(r['í‰ì ']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['ì´ë¦„']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">ì¹´í…Œê³ ë¦¬: {r['ì¹´í…Œê³ ë¦¬']}</p>
            <p style="margin:0;font-size:12px;color:#555;">â­ í‰ì : {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['ì£¼ì†Œ']}</p>
        </div>
        """
    return html

# ------------ ë©”ì¸: ì§€ë„ ìƒì„± & í‘œì‹œ ------------
def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000

    # 1) í‰ì  ì—†ì´ ì¥ì†Œ ëª©ë¡ ë¨¼ì € ìˆ˜ì§‘
    dfs = []
    cats = ["í—¬ìŠ¤ì¥", "ê³µì›", "ë‚´ê³¼", "ì¢…í•©ë³‘ì›", "ë‚´ë¶„ë¹„ë‚´ê³¼"] if category == "ì „ì²´" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius)
        if not d.empty:
            dfs.append(d)

    # ì§€ë„ì˜ ê¸°ì¤€ ì¢Œí‘œ
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780

    # ì§€ë„ ìƒì„±
    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>ë‚´ ìœ„ì¹˜</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

    # 2) í•­ìƒ í‰ì  ìˆ˜ì§‘í•´ì„œ í‘œì‹œ (ì •ë ¬ì€ ì•„ë˜ì—ì„œ ì„ íƒì ìœ¼ë¡œ ì ìš©)
    if not df.empty:
        driver = init_driver()
        try:
            ratings = get_kakao_ratings_bulk(driver, df["ë§í¬"].tolist())
            df["í‰ì "] = df["ë§í¬"].map(ratings)
        finally:
            try: driver.quit()
            except: pass

        # 3) ì •ë ¬
        if sort == "rating" and "í‰ì " in df.columns:
            df = df.sort_values("í‰ì ", ascending=False, na_position="last")

        # ë§ˆì»¤ ì¶”ê°€
        for row in df.itertuples():
            icon_url = icon_urls.get(row.ì¹´í…Œê³ ë¦¬)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.ëŒ€í‘œì´ë¯¸ì§€}' width='150' style='border-radius:6px;object-fit:cover;'><br>" if getattr(row, "ëŒ€í‘œì´ë¯¸ì§€", None) else ""
            rating_txt = "ì •ë³´ ì—†ìŒ"
            val = getattr(row, 'í‰ì ', None)
            if pd.notna(val):
                try:
                    rating_txt = f"{float(val):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.ì´ë¦„}</b><br>"
                f"ì¹´í…Œê³ ë¦¬: {row.ì¹´í…Œê³ ë¦¬}<br>"
                f"ì£¼ì†Œ: {row.ì£¼ì†Œ}<br>"
                f"ì „í™”: {row.ì „í™”ë²ˆí˜¸}<br>"
                f"â­ í‰ì : {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.ë§í¬}' target='_blank'>ì¹´ì¹´ì˜¤ë§µì—ì„œ ë³´ê¸°</a>",
                max_width=320
            )
            folium.Marker([row.ìœ„ë„, row.ê²½ë„], icon=icon, popup=popup).add_to(m)

    # 4) ì €ì¥ + ë Œë”
    os.makedirs("./_maps", exist_ok=True)
    map_path = f"./_maps/folium_map_{int(time.time())}.html"
    m.save(map_path)

    cards_html = top5_cards_html(df)
    full_html = f"""
    <div style="display:flex;gap:20px;align-items:flex-start;">
      <div style="flex:2;min-width:500px;">
        <iframe src="{map_path}" style="width:100%;height:600px;border:none;"></iframe>
      </div>
      <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
        {cards_html}
      </div>
    </div>
    """
    return full_html

# ------------ ìœ„ì ¯ UI ------------
address_input = Text(value="ì„œìš¸ ì¢…ë¡œêµ¬ ì„¸ì¢…ëŒ€ë¡œ 110", description="ì£¼ì†Œ:", layout={'width':'60%'})
category_dd   = Dropdown(options=["ì „ì²´","í—¬ìŠ¤ì¥","ê³µì›","ë‚´ê³¼","ì¢…í•©ë³‘ì›","ë‚´ë¶„ë¹„ë‚´ê³¼"], value="ì „ì²´", description="ë³´ê¸°:")
sort_dd       = Dropdown(options=[("ì •í™•ë„ìˆœ","accuracy"),("ê°€ê¹Œìš´ìˆœ","distance"),("í‰ì ìˆœ","rating")], value="accuracy", description="ì •ë ¬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="ë°˜ê²½(km):")
btn_search    = Button(description="ê²€ìƒ‰", button_style="success")
out           = Output()

def on_click(b):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
btn_search.on_click(on_click)
display(ui)
```

```python
import re
import folium
import pandas as pd
import requests, os, time
from typing import Optional
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# ====== ì„¤ì • ======
# ì§€ë„ ë Œë” ë°©ì‹: "inline"(ê¶Œì¥) ë˜ëŠ” "iframe"(íŒŒì¼ ì €ì¥ í›„ iframe)
RENDER_MODE = "inline"

# === API KEYS ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# -------------------------------
# (A) ë„¤ì´ë²„ ì´ë¯¸ì§€ APIë§Œ ì‚¬ìš©í•´ ëŒ€í‘œì´ë¯¸ì§€ ì„ íƒ
# -------------------------------
def _normalize(s: str) -> str:
    if not isinstance(s, str):
        return ""
    return re.sub(r"[\s\-\_\|\(\)\[\]{}Â·â€¢âˆ™,.;:!?'\"<>@#%^&*+=/\\]", "", s.lower())

def _extract_region_token(addr_text: str) -> str:
    """ì£¼ì†Œì—ì„œ ìš°ì„ ìˆœìœ„: ë™ > êµ¬ > ì > ë©´ í•œ ê°œ í† í°ë§Œ ì¶”ì¶œ"""
    if not isinstance(addr_text, str):
        return ""
    toks = [t for t in addr_text.split() if t]
    for suf in ("ë™", "êµ¬", "ì", "ë©´"):
        cand = [t for t in toks if t.endswith(suf)]
        if cand:
            return cand[0]
    return ""

def _is_good_ext(link: str) -> bool:
    base = link.lower().split("?")[0]
    return base.endswith((".jpg", ".jpeg", ".png", ".webp"))

def _score_naver_item(item, name_norm: str, region_norm: str) -> int:
    """ì œëª©/ë§í¬ í† í° ë§¤ì¹­ + í•´ìƒë„ ê°€ì , ìŒì‹/ê´‘ê³ ë¥˜ëŠ” ì¦‰ì‹œ íƒˆë½"""
    title = (item.get("title") or "")
    link  = (item.get("link") or "")
    w = int(item.get("sizewidth", 0) or 0)
    h = int(item.get("sizeheight", 0) or 0)

    if not link.startswith("http") or not _is_good_ext(link):
        return -999

    low_title = title.lower()
    low_link  = link.lower()
    # ìŒì‹/ë§›ì§‘/ê´‘ê³ ì„± í‚¤ì›Œë“œ ë°°ì œ
    bad = ("ì¹˜í‚¨","ë§›ì§‘","ì‹ë‹¹","ì¹´í˜","ë©”ë‰´","ë°°ë‹¬","ë¶„ì‹","ìˆ ì§‘","ë§¥ì£¼",
           "dessert","bread","recipe","food","mangoplate","siksin","coupon","banner","ad")
    if any(b in low_title or b in low_link for b in bad):
        return -999

    score = 0
    t_norm = _normalize(title)
    l_norm = _normalize(link)
    if name_norm and (name_norm in t_norm or name_norm in l_norm):
        score += 3
    if region_norm and (region_norm in t_norm or region_norm in l_norm):
        score += 2
    if w >= 400 and h >= 300:
        score += 1
    return score

def get_naver_image_only(place_name: str, addr_text: str) -> Optional[str]:
    """
    ë„¤ì´ë²„ ì´ë¯¸ì§€ OpenAPIë§Œ ì‚¬ìš©.
    ì¿¼ë¦¬: '{ë™/êµ¬} {ìƒí˜¸ëª…} í—¬ìŠ¤/ì‹œì„¤ ê´€ë ¨ ì–‘ì˜ í‚¤ì›Œë“œ -ìŒì‹/ë§›ì§‘ ë¶€ì • í‚¤ì›Œë“œ'
    ìƒìœ„ 10ê°œ ì¤‘ ìŠ¤ì½”ì–´ê°€ ê°€ì¥ ë†’ì€ ë§í¬ë¥¼ ëŒ€í‘œì´ë¯¸ì§€ë¡œ ì„ íƒ(ìŠ¤ì½”ì–´<2ë©´ None).
    """
    region = _extract_region_token(addr_text)
    # ì¥ì†Œ/ì‹œì„¤ ìœ„ì£¼ë¡œ ë³´ì´ê²Œ ê¸ì • í‚¤ì›Œë“œ, ìŒì‹/ë§›ì§‘ ì œê±°
    positive = "í—¬ìŠ¤ì¥ ì²´ìœ¡ê´€ ì‹¤ë‚´ gym fitness ì‹œì„¤ ë‚´ë¶€"
    negative = "-ì¹˜í‚¨ -ë§›ì§‘ -ì‹ë‹¹ -ì¹´í˜ -ë©”ë‰´ -ë°°ë‹¬ -ë¶„ì‹ -ìˆ ì§‘ -ë§¥ì£¼ -dessert -bread -recipe -food -mangoplate -siksin -coupon -banner -ad"
    base_query = f"{region} {place_name}".strip() if region else f"{place_name} {addr_text}".strip()
    query = f"{base_query} {positive} {negative}"

    url = "https://openapi.naver.com/v1/search/image"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }
    params = {"query": query, "display": 10, "sort": "sim"}

    try:
        r = requests.get(url, headers=headers, params=params, timeout=8)
        r.raise_for_status()
        items = r.json().get("items", []) or []
        if not items:
            return None
        name_norm   = _normalize(place_name)
        region_norm = _normalize(region)
        best, best_score = None, -999
        for it in items:
            sc = _score_naver_item(it, name_norm, region_norm)
            if sc > best_score:
                best, best_score = it, sc
        if best and best_score >= 2:
            return best.get("link")
        return None
    except:
        return None

# -----------------------------------
# (B) ì¹´ì¹´ì˜¤ ì¢Œí‘œ ë³€í™˜
# -----------------------------------
def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) ì£¼ì†Œ ê²€ìƒ‰
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) í‚¤ì›Œë“œ
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) ìˆ«ì ì œê±° í›„ í‚¤ì›Œë“œ
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

# -------------------------------
# (C) Selenium í‰ì  í¬ë¡¤ë§
# -------------------------------
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                    )
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

# ------------------------------------------
# (D) ì¹´ì¹´ì˜¤ APIë¡œ ì¥ì†Œ ê²€ìƒ‰ + ëŒ€í‘œì´ë¯¸ì§€(ë„¤ì´ë²„ë§Œ) / í‰ì 
# ------------------------------------------
def get_places(category, address, size=15, sort="accuracy", radius=3000):
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_name = p.get("place_name")
        addr_q     = p.get("road_address_name") or p.get("address_name")
        place_url  = p.get("place_url")

        # âœ… ëŒ€í‘œì´ë¯¸ì§€: ë„¤ì´ë²„ ì´ë¯¸ì§€ APIë§Œ ì‚¬ìš©
        img = get_naver_image_only(place_name, addr_q)

        rows.append({
            "ì´ë¦„": place_name,
            "ì¹´í…Œê³ ë¦¬": category,
            "ì£¼ì†Œ": addr_q,
            "ìœ„ë„": float(p["y"]),
            "ê²½ë„": float(p["x"]),
            "ì „í™”ë²ˆí˜¸": p.get("phone"),
            "ë§í¬": place_url,
            "ëŒ€í‘œì´ë¯¸ì§€": img
        })
    return pd.DataFrame(rows)

# -------------------
# (E) ì•„ì´ì½˜ & ì¹´ë“œ
# -------------------
icon_urls = {
    "í—¬ìŠ¤ì¥": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "ê³µì›": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "ë‚´ê³¼": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "ì¢…í•©ë³‘ì›": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "ë‚´ë¶„ë¹„ë‚´ê³¼": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty:
        return "<p style='color:#777'>í‘œì‹œí•  ê²°ê³¼ê°€ ì—†ìŠµë‹ˆë‹¤.</p>"
    top5 = df.head(5)
    html = "<h3>ğŸ“Œ Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['ëŒ€í‘œì´ë¯¸ì§€']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" \
                  if pd.notna(r.get("ëŒ€í‘œì´ë¯¸ì§€")) and r.get("ëŒ€í‘œì´ë¯¸ì§€") else ""
        rating_txt = "ì •ë³´ ì—†ìŒ"
        if "í‰ì " in r and pd.notna(r["í‰ì "]):
            try:
                rating_txt = f"{float(r['í‰ì ']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['ì´ë¦„']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">ì¹´í…Œê³ ë¦¬: {r['ì¹´í…Œê³ ë¦¬']}</p>
            <p style="margin:0;font-size:12px;color:#555;">â­ í‰ì : {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['ì£¼ì†Œ']}</p>
        </div>
        """
    return html

# --------------------------------
# (F) ì§€ë„ ìƒì„± & í‘œì‹œ (ìë™ ê°±ì‹ /ì¹´í…Œê³ ë¦¬ í™•ì‹¤ í•„í„°)
# --------------------------------
def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000

    # ë°ì´í„° ìˆ˜ì§‘ (ì´ë¯¸ì§€ í¬í•¨)
    dfs = []
    cats = ["í—¬ìŠ¤ì¥", "ê³µì›", "ë‚´ê³¼", "ì¢…í•©ë³‘ì›", "ë‚´ë¶„ë¹„ë‚´ê³¼"] if category == "ì „ì²´" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius)
        if not d.empty:
            dfs.append(d)

    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780  # fallback

    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>ë‚´ ìœ„ì¹˜</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

    # âœ… ìµœì¢… ì•ˆì „ì¥ì¹˜: ì„ íƒ ì¹´í…Œê³ ë¦¬ë§Œ ë‚¨ê¸°ê¸°
    if not df.empty and category != "ì „ì²´":
        df = df[df["ì¹´í…Œê³ ë¦¬"] == category].copy()

    # í‰ì ì€ í•­ìƒ ìˆ˜ì§‘í•´ì„œ í‘œì‹œ, 'í‰ì ìˆœ'ì¼ ë•Œë§Œ ì •ë ¬
    if not df.empty:
        driver = init_driver()
        try:
            ratings = get_kakao_ratings_bulk(driver, df["ë§í¬"].tolist())
            df["í‰ì "] = df["ë§í¬"].map(ratings)
        finally:
            try: driver.quit()
            except: pass

        if sort == "rating" and "í‰ì " in df.columns:
            df = df.sort_values("í‰ì ", ascending=False, na_position="last")

        for row in df.itertuples():
            icon_url = icon_urls.get(row.ì¹´í…Œê³ ë¦¬)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.ëŒ€í‘œì´ë¯¸ì§€}' width='150' style='border-radius:6px;object-fit:cover;'><br>" \
                       if getattr(row, "ëŒ€í‘œì´ë¯¸ì§€", None) else ""
            rating_txt = "ì •ë³´ ì—†ìŒ"
            val = getattr(row, 'í‰ì ', None)
            if pd.notna(val):
                try:
                    rating_txt = f"{float(val):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.ì´ë¦„}</b><br>"
                f"ì¹´í…Œê³ ë¦¬: {row.ì¹´í…Œê³ ë¦¬}<br>"
                f"ì£¼ì†Œ: {row.ì£¼ì†Œ}<br>"
                f"ì „í™”: {row.ì „í™”ë²ˆí˜¸ or '-'}<br>"
                f"â­ í‰ì : {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.ë§í¬}' target='_blank' rel='noopener noreferrer'>ì¹´ì¹´ì˜¤ë§µì—ì„œ ë³´ê¸°</a>",
                max_width=320, parse_html=True
            )
            folium.Marker([row.ìœ„ë„, row.ê²½ë„], icon=icon, popup=popup).add_to(m)

    # ë Œë”
    cards_html = top5_cards_html(df)
    if RENDER_MODE == "iframe":
        os.makedirs("./_maps", exist_ok=True)
        map_path = f"./_maps/folium_map_{int(time.time())}.html"
        m.save(map_path)
        full_html = f"""
        <div style="display:flex;gap:20px;align-items:flex-start;">
          <div style="flex:2;min-width:500px;">
            <iframe src="{map_path}"
                    style="width:100%;height:600px;border:none;"
                    sandbox="allow-scripts allow-popups allow-popups-to-escape-sandbox allow-top-navigation-by-user-activation">
            </iframe>
          </div>
          <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
            {cards_html}
          </div>
        </div>
        """
        return full_html
    else:
        map_html = m._repr_html_()
        full_html = f"""
        <div style="display:flex;gap:20px;align-items:flex-start;">
          <div style="flex:2;min-width:500px;">{map_html}</div>
          <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
            {cards_html}
          </div>
        </div>
        """
        return full_html

# ------------ ìœ„ì ¯ UI (ìë™ ê°±ì‹ ) ------------
address_input = Text(value="ëŒ€êµ¬ì‹œ ë‹¬ì„œêµ¬ ì›”ì„±ë™ ì›”ì„±eí¸í•œì„¸ìƒ", description="ì£¼ì†Œ:", layout={'width':'60%'})
category_dd   = Dropdown(options=["ì „ì²´","í—¬ìŠ¤ì¥","ê³µì›","ë‚´ê³¼","ì¢…í•©ë³‘ì›","ë‚´ë¶„ë¹„ë‚´ê³¼"], value="ì „ì²´", description="ë³´ê¸°:")
sort_dd       = Dropdown(options=[("ì •í™•ë„ìˆœ","accuracy"),("ê°€ê¹Œìš´ìˆœ","distance"),("í‰ì ìˆœ","rating")], value="accuracy", description="ì •ë ¬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="ë°˜ê²½(km):")
btn_search    = Button(description="ê²€ìƒ‰", button_style="success")
out           = Output()

def run_search(*args):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

# ë²„íŠ¼ í´ë¦­
btn_search.on_click(lambda b: run_search())

# âœ… ìœ„ì ¯ ë³€ê²½ ì¦‰ì‹œ ìë™ ì—…ë°ì´íŠ¸
category_dd.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
sort_dd.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
radius_slider.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
address_input.observe(lambda change: run_search() if change['name']=='value' else None, names='value')

# UI ë Œë” + ìµœì´ˆ 1íšŒ ì‹¤í–‰
ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
display(ui)
run_search()
```
