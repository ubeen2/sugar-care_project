```python
import re
import folium
import pandas as pd
import requests, os, time
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# === API KEYS (교체 권장) ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# ------------ 유틸 ------------
def get_naver_image(query):
    """1순위 이미지 소스: 네이버 이미지 API (실패 시 None 반환)"""
    url = "https://openapi.naver.com/v1/search/image"
    headers = {"X-Naver-Client-Id": NAVER_CLIENT_ID,
               "X-Naver-Client-Secret": NAVER_CLIENT_SECRET}
    params = {"query": query, "display": 1, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        res = r.json()
        if res.get("items"):
            link = res["items"][0]["link"]
            if isinstance(link, str) and link.startswith("http"):
                return link
    except:
        pass
    return None

def get_og_image_from_kakao(place_url):
    """2순위(폴백) 이미지: 카카오 장소 og:image"""
    try:
        h = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
        }
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']', html, re.IGNORECASE)
        if m:
            return m.group(1)
    except:
        return None
    return None

def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) 주소 검색
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) 키워드 그대로
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) 숫자 제거 후 키워드
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    # 탐지 회피 & UA 지정
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    """문자열에서 첫 번째 실수형 숫자 추출 (예: '평점 4.8/5' -> 4.8)"""
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    """요청: 평점은 Selenium으로 가져오기"""
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                    )
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

def get_places(category, address, size=15, sort="accuracy", radius=3000, driver=None):
    """사진 우선순위: 네이버 API → (없으면) 카카오 og:image"""
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_url = p.get("place_url")
        addr_q = p["road_address_name"] or p["address_name"]
        # 1순위: 네이버 이미지
        img = get_naver_image(f"{p['place_name']} {addr_q}")
        # 2순위: 카카오 og:image
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "이름": p["place_name"],
            "카테고리": category,
            "주소": addr_q,
            "위도": float(p["y"]),
            "경도": float(p["x"]),
            "전화번호": p["phone"],
            "링크": place_url,
            "대표이미지": img  # None이면 카드/팝업에서 이미지 미노출
        })
    df = pd.DataFrame(rows)

    # 요청: 평점은 셀레니움으로
    if sort == "rating" and driver is not None and not df.empty:
        ratings = get_kakao_ratings_bulk(driver, df["링크"].tolist())
        df["평점"] = df["링크"].map(ratings)
        if df["평점"].notna().sum() > 0:
            df = df.sort_values("평점", ascending=False, na_position="last")

    return df

icon_urls = {
    "헬스장": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "공원": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "내과": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "종합병원": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "내분비내과": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty: 
        return "<p style='color:#777'>표시할 결과가 없습니다.</p>"
    top5 = df.head(5)
    html = "<h3>📌 Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['대표이미지']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" \
                  if pd.notna(r.get("대표이미지")) and r.get("대표이미지") else ""
        rating_txt = "정보 없음"
        if "평점" in r and pd.notna(r["평점"]):
            try:
                rating_txt = f"{float(r['평점']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['이름']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">카테고리: {r['카테고리']}</p>
            <p style="margin:0;font-size:12px;color:#555;">⭐ 평점: {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['주소']}</p>
        </div>
        """
    return html

def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000
    driver = init_driver() if sort == "rating" else None

    # 데이터 수집
    dfs = []
    cats = ["헬스장", "공원", "내과", "종합병원", "내분비내과"] if category == "전체" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius, driver=driver)
        if not d.empty:
            dfs.append(d)
    if driver:
        try:
            driver.quit()
        except:
            pass

    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780  # 서울시청 fallback

    # 지도 생성 (iframe으로만 출력해 중복 지도 방지)
    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>내 위치</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    if dfs:
        df = pd.concat(dfs, ignore_index=True)
        for row in df.itertuples():
            icon_url = icon_urls.get(row.카테고리)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.대표이미지}' width='150' style='border-radius:6px;object-fit:cover;'><br>" \
                       if getattr(row, "대표이미지", None) else ""
            rating_txt = "정보 없음"
            if hasattr(row, "평점") and pd.notna(getattr(row, "평점")):
                try:
                    rating_txt = f"{float(getattr(row, '평점')):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.이름}</b><br>"
                f"카테고리: {row.카테고리}<br>"
                f"주소: {row.주소}<br>"
                f"전화: {row.전화번호}<br>"
                f"⭐ 평점: {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.링크}' target='_blank'>카카오맵에서 보기</a>",
                max_width=320
            )
            folium.Marker([row.위도, row.경도], icon=icon, popup=popup).add_to(m)
    else:
        df = pd.DataFrame()

    # 지도를 파일로 저장하고 iframe으로 표시
    os.makedirs("./_maps", exist_ok=True)
    map_path = f"./_maps/folium_map_{int(time.time())}.html"
    m.save(map_path)

    cards_html = top5_cards_html(df)
    full_html = f"""
    <div style="display:flex;gap:20px;align-items:flex-start;">
      <div style="flex:2;min-width:500px;">
        <iframe src="{map_path}" style="width:100%;height:600px;border:none;"></iframe>
      </div>
      <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
        {cards_html}
      </div>
    </div>
    """
    return full_html

# ------------ 위젯 UI ------------
address_input = Text(value="서울 종로구 세종대로 110", description="주소:", layout={'width':'60%'})
category_dd   = Dropdown(options=["전체","헬스장","공원","내과","종합병원","내분비내과"], value="전체", description="보기:")
# 라벨은 '가까운순', 값은 'distance'
sort_dd       = Dropdown(options=[("정확도순","accuracy"),("가까운순","distance"),("평점순","rating")], value="accuracy", description="정렬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="반경(km):")
btn_search    = Button(description="검색", button_style="success")
out           = Output()

def on_click(b):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
btn_search.on_click(on_click)
display(ui)
```

```python
import folium
import pandas as pd
import requests
from ipywidgets import interact, Text, Dropdown, IntSlider
from IPython.display import display, HTML

# 🔑 API 키
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# ✅ 네이버 이미지 검색 API
def get_naver_image(query):
    url = "https://openapi.naver.com/v1/search/image"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }
    params = {"query": query, "display": 1, "sort": "sim"}
    res = requests.get(url, headers=headers, params=params)
    try:
        data = res.json()
        if "items" in data and data["items"]:
            return data["items"][0]["link"]
    except:
        print("❌ 네이버 이미지 API 에러:", res.text)
    return "https://via.placeholder.com/150"

# ✅ 주소 → 좌표 변환 (자동 보정)
def get_coords_from_address(address):
    # 1차: 주소 검색 API
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {"query": address}
    res = requests.get(url, headers=headers, params=params)
    data = res.json()

    if "documents" in data and data["documents"]:
        y = float(data["documents"][0]["y"])
        x = float(data["documents"][0]["x"])
        return y, x
    
    # 2차: 키워드 검색 API (공백 제거, 상세주소 → 키워드 처리)
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    params = {"query": address.replace(" ", ""), "size": 1}
    res = requests.get(url, headers=headers, params=params)
    data = res.json()
    if "documents" in data and data["documents"]:
        y = float(data["documents"][0]["y"])
        x = float(data["documents"][0]["x"])
        return y, x

    return None, None

# ✅ 카카오 API로 장소 검색
def get_places(category, address, size=5, sort="accuracy", radius=3000):
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        print(f"⚠️ [{address}] 좌표를 가져오지 못했습니다.")
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "sort": "distance" if sort == "distance" else "accuracy",
        "x": lon,
        "y": lat,
        "radius": radius
    }
    res = requests.get(url, headers=headers, params=params)
    data = res.json()

    if "documents" not in data or not data["documents"]:
        print(f"⚠️ {address} {category} 검색 결과 없음")
        return pd.DataFrame()

    docs = data["documents"]

    df = pd.DataFrame([
        {
            "이름": place["place_name"],
            "카테고리": category,
            "주소": place["road_address_name"] or place["address_name"],
            "위도": float(place["y"]),
            "경도": float(place["x"]),
            "전화번호": place["phone"],
            "링크": place["place_url"],
            "대표이미지": get_naver_image(f"{address} {place['place_name']}")
        }
        for place in docs
    ])

    return df

# ✅ 아이콘 설정
icon_urls = {
    "헬스장": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "공원": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "내과": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "종합병원": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "내분비내과": "https://img.icons8.com/ios/50/pills.png"
}

# ✅ Top5 카드 HTML
def generate_top5_cards_html(df, sort):
    if sort not in ["distance"] or df.empty:
        return ""
    top5 = df.head(5)
    cards_html = "<h3>📌 Top 5</h3>"
    for _, row in top5.iterrows():
        cards_html += f"""
        <div style='width:220px; border:1px solid #ddd; border-radius:10px; padding:10px; margin-bottom:10px; box-shadow:2px 2px 5px #ccc;'>
            <img src='{row['대표이미지']}' style='width:100%; height:120px; object-fit:cover; border-radius:8px;'>
            <h4 style='margin:8px 0 4px 0; font-size:14px;'>{row['이름']}</h4>
            <p style='margin:0; font-size:12px; color:#555;'>카테고리: {row['카테고리']}</p>
            <p style='margin:0; font-size:11px; color:#777;'>{row['주소']}</p>
        </div>
        """
    return cards_html

# ✅ 지도 생성 함수
def make_map(address="서울 종로구", category="전체", sort="accuracy", radius_km=3):
    radius = radius_km * 1000  # km → m 변환
    df_list = []
    for cat in ["헬스장", "공원", "내과", "종합병원", "내분비내과"]:
        if category == "전체" or category == cat:
            df_tmp = get_places(cat, address, size=5, sort=sort, radius=radius)
            if not df_tmp.empty:
                df_list.append(df_tmp)

    if not df_list:
        print("⚠️ 선택한 조건에 맞는 장소가 없습니다.")
        return folium.Map(location=[37.5665, 126.9780], zoom_start=12)

    df = pd.concat(df_list, ignore_index=True)

    # 지도
    m = folium.Map(location=[df["위도"].mean(), df["경도"].mean()], zoom_start=13)

    for row in df.itertuples():
        icon_url = icon_urls.get(row.카테고리)
        icon = folium.CustomIcon(icon_url, icon_size=(35, 35))
        popup = folium.Popup(
            f"<b>{row.이름}</b><br>"
            f"카테고리: {row.카테고리}<br>"
            f"주소: {row.주소}<br>"
            f"전화: {row.전화번호}<br>"
            f"<img src='{row.대표이미지}' width='150'><br>"
            f"<a href='{row.링크}' target='_blank'>카카오맵에서 보기</a>",
            max_width=300
        )
        folium.Marker([row.위도, row.경도], icon=icon, popup=popup).add_to(m)

    # 지도 + 카드
    map_html = m._repr_html_()
    cards_html = generate_top5_cards_html(df, sort)

    full_html = f"""
    <div style="display:flex; gap:20px;">
        <div style="flex:2;">{map_html}</div>
        <div style="flex:1; max-width:250px; overflow-y:auto;">{cards_html}</div>
    </div>
    """
    display(HTML(full_html))

# ✅ 인터랙티브 UI
interact(
    make_map,
    address=Text(value="서울 종로구 세종대로 110", description="주소:"),
    category=Dropdown(options=["전체", "헬스장", "공원", "내과", "종합병원", "내분비내과"], value="전체", description="보기:"),
    sort=Dropdown(options=[("정확도순", "accuracy"), ("거리순", "distance")], value="accuracy", description="정렬:"),
    radius_km=IntSlider(value=3, min=1, max=20, step=1, description="반경(km):")
)
```

```python
# -*- coding: utf-8 -*-
import re, time
import folium
import pandas as pd
import requests

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Jupyter widgets
from ipywidgets import interact, Text, Dropdown

# =========================
# 🔑 API KEYS (채워 넣기)
# =========================
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# =========================
# 🔧 유틸
# =========================
def init_driver():
    """Selenium Chrome Headless 초기화(탐지 회피 옵션 포함)"""
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")           # 최신 헤드리스
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    # 탐지 회피 & UA 고정
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
    )
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text: str):
    """문자열에서 첫 번째 실수형 숫자 추출 (예: '평점 4.8/5' -> 4.8)"""
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    return float(m.group(1)) if m else None

def get_kakao_ratings_bulk(driver, place_urls):
    """
    ✅ 카카오 장소 상세 페이지들에서 평점을 일괄 크롤링
    - 동일 드라이버로 순회 → 속도/안정성 ↑
    """
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            time.sleep(0.25)

            # 최대 3회 재시도 (지연렌더 대비)
            for _ in range(3):
                for sel in selectors:
                    try:
                        elem = WebDriverWait(driver, 3).until(
                            EC.presence_of_element_located((By.CSS_SELECTOR, sel))
                        )
                        rating = _extract_first_float((elem.text or "").strip())
                        if rating is not None:
                            break
                    except Exception:
                        continue
                if rating is not None:
                    break
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(0.35)
        except Exception as e:
            # 디버깅용 로그(필요시 유지)
            # print("평점 크롤링 실패:", e)
            rating = None
        results[url] = rating
    return results

def get_naver_image(query):
    """
    ✅ 1순위 이미지: 네이버 이미지 검색 API
    - 가장 유사한 1장 반환, 실패 시 None
    """
    url = "https://openapi.naver.com/v1/search/image"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }
    params = {"query": query, "display": 1, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        data = r.json()
        if data.get("items"):
            link = data["items"][0].get("link")
            if isinstance(link, str) and link.startswith("http"):
                return link
    except Exception:
        pass
    return None

def get_og_image_from_kakao(place_url):
    """
    2순위(폴백) 이미지: 카카오 장소 상세 페이지의 og:image
    """
    try:
        h = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                          "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
        }
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(
            r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']',
            html, re.IGNORECASE
        )
        if m:
            return m.group(1)
    except Exception:
        return None
    return None

# =========================
# 🗺 장소 검색 (카카오 API)
# =========================
def get_places(category, query="종로구", size=5, sort="accuracy", driver=None, fetch_ratings=True):
    """
    카카오 키워드 검색
    - sort: accuracy | distance (카카오 API에서 지원)
    - 사진: 네이버 API 우선 → 카카오 og:image 폴백
    - 평점: fetch_ratings=True일 때 Selenium 일괄 크롤링
    """
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": f"{query} {category}",
        "size": size,
        "sort": sort if sort in ["accuracy", "distance"] else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7)
        res.raise_for_status()
        data = res.json()
    except Exception as e:
        print("❌ 카카오 API 요청 실패:", e)
        return pd.DataFrame()

    docs = data.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_url = p.get("place_url")
        addr_q = p.get("road_address_name") or p.get("address_name") or ""

        # ✅ 사진: 네이버 → 카카오 og:image
        img = get_naver_image(f"{p.get('place_name','')} {addr_q}")
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "이름": p.get("place_name"),
            "카테고리": category,
            "주소": addr_q,
            "위도": float(p.get("y")),
            "경도": float(p.get("x")),
            "전화번호": p.get("phone"),
            "링크": place_url,
            "대표이미지": img
        })

    df = pd.DataFrame(rows)

    # ✅ 평점: 요청대로 Selenium으로 (대량)
    if fetch_ratings and driver is not None and not df.empty:
        ratings = get_kakao_ratings_bulk(driver, df["링크"].tolist())
        df["평점"] = df["링크"].map(ratings)

    return df

# =========================
# 🖼 마커 아이콘
# =========================
icon_urls = {
    "헬스장": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "공원": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "내과": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "종합병원": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "내분비내과": "https://img.icons8.com/ios/50/pills.png"
}

# =========================
# 🗺 지도 생성
# =========================
def make_map(query="대구 북구", category="전체", sort="accuracy"):
    """
    - category: '전체'면 5개 카테고리 모두 수집 후 합침
    - sort == 'rating'이면 평점 기준 정렬
    """
    cats = ["헬스장", "공원", "내과", "종합병원", "내분비내과"] if category == "전체" else [category]

    # ✅ 드라이버를 단 한 번만 열어 모든 평점을 일괄 수집
    driver = init_driver()
    try:
        dfs = []
        for cat in cats:
            # 평점은 항상 표시되길 원하므로 fetch_ratings=True 유지
            df_tmp = get_places(cat, query=query, size=5, sort=("distance" if sort=="distance" else "accuracy"),
                                driver=driver, fetch_ratings=True)
            if not df_tmp.empty:
                dfs.append(df_tmp)
    finally:
        try:
            driver.quit()
        except:
            pass

    if not dfs:
        print("⚠️ 선택한 조건에 맞는 장소가 없습니다.")
        return folium.Map(location=[37.5665, 126.9780], zoom_start=12)

    df = pd.concat(dfs, ignore_index=True)

    # ✅ 평점순 정렬 요청 시
    if sort == "rating" and "평점" in df.columns:
        df = df.sort_values("평점", ascending=False, na_position="last")

    # 지도 중심: 평균 좌표
    m = folium.Map(location=[df["위도"].mean(), df["경도"].mean()], zoom_start=13)

    # 마커 생성
    for row in df.itertuples():
        if category == "전체" or row.카테고리 == category:
            icon_url = icon_urls.get(row.카테고리)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None

            rating_txt = "정보 없음"
            if hasattr(row, "평점") and pd.notna(getattr(row, "평점")):
                try:
                    rating_txt = f"{float(getattr(row, '평점')):.1f}"
                except Exception:
                    pass

            img_html = ""
            if getattr(row, "대표이미지", None):
                img_html = f"<br><img src='{row.대표이미지}' width='160' style='border-radius:6px;object-fit:cover;'>"

            popup = folium.Popup(
                f"<b>{row.이름}</b><br>"
                f"카테고리: {row.카테고리}<br>"
                f"주소: {row.주소}<br>"
                f"전화: {row.전화번호 or '-'}<br>"
                f"⭐ 평점: {rating_txt}"
                f"{img_html}"
                f"<br><a href='{row.링크}' target='_blank'>카카오맵에서 보기</a>",
                max_width=320
            )
            folium.Marker([row.위도, row.경도], icon=icon, popup=popup).add_to(m)

    return m

# =========================
# 🎛 인터랙티브 UI
# =========================
interact(
    make_map,
    query=Text(value="대구 북구", description="지역:"),
    category=Dropdown(
        options=["전체", "헬스장", "공원", "내과", "종합병원", "내분비내과"],
        value="전체",
        description="보기:"
    ),
    sort=Dropdown(
        options=[("정확도순", "accuracy"), ("가까운순", "distance"), ("평점순", "rating")],
        value="accuracy",
        description="정렬:"
    )
)
```

```python
import re
import folium
import pandas as pd
import requests, os, time
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# === API KEYS (교체 권장) ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# ------------ 공통 유틸 ------------
def _normalize(s: str) -> str:
    if not isinstance(s, str):
        return ""
    return re.sub(r"[\s\-\_\|\(\)\[\]{}·•∙,.;:!?'\"<>@#%^&*+=/\\]", "", s.lower())

def extract_region_token(addr_text: str) -> str:
    """주소에서 검색용 토큰 추출: 우선순위 '동' > '구' > '읍' > '면'"""
    if not isinstance(addr_text, str):
        return ""
    toks = [t for t in addr_text.split() if t]
    dong = [t for t in toks if t.endswith("동")]
    gu   = [t for t in toks if t.endswith("구")]
    eup  = [t for t in toks if t.endswith("읍")]
    myeon= [t for t in toks if t.endswith("면")]
    return (dong or gu or eup or myeon or [""])[0]

# ------------ 이미지 유틸 (네이버 엄격 필터) ------------
def _is_likely_photo(link: str, w: int=None, h: int=None) -> bool:
    if not isinstance(link, str) or not link.startswith("http"):
        return False
    base = link.lower().split("?")[0]
    # 확장자 필터
    if not base.endswith((".jpg", ".jpeg", ".png", ".webp")):
        return False
    # 너무 작은 썸네일 제외
    try:
        if w is not None and h is not None and (w < 200 or h < 150):
            return False
    except:
        pass
    # 금지 키워드(아이콘/로고/지도/배너/메뉴판 등)
    bad = ("emoji","icon","favicon","sprite","logo","map","kakaomap","navermap","banner","menu","price","coupon","ad")
    if any(k in base for k in bad):
        return False
    return True

def _score_naver_item(item, name_norm: str, region_norm: str) -> int:
    title = item.get("title") or ""
    link  = item.get("link") or ""
    w = int(item.get("sizewidth", 0)) if item.get("sizewidth") else None
    h = int(item.get("sizeheight", 0)) if item.get("sizeheight") else None

    if not _is_likely_photo(link, w, h):
        return -999

    t_norm = _normalize(title)
    l_norm = _normalize(link)
    score = 0
    # 이름/지역 토큰 매칭 가중치
    if name_norm and (name_norm in t_norm or name_norm in l_norm):
        score += 3
    if region_norm and (region_norm in t_norm or region_norm in l_norm):
        score += 2
    # 해상도 가점(대략)
    if (w or 0) >= 400 and (h or 0) >= 300:
        score += 1
    return score

def get_naver_image_strict(place_name: str, addr_text: str):
    """
    네이버 이미지 API만 사용하되, 상위 10개 중 가장 관련성 높은 이미지만 선택.
    - 쿼리는 '{동/구} {상호명}' (토큰 없으면 '상호명 주소')
    - 점수 < 2 이면 None 반환 → 이후 og:image 폴백
    """
    region = extract_region_token(addr_text)
    query = f"{region} {place_name}" if region else f"{place_name} {addr_text}"

    url = "https://openapi.naver.com/v1/search/image"
    headers = {"X-Naver-Client-Id": NAVER_CLIENT_ID,
               "X-Naver-Client-Secret": NAVER_CLIENT_SECRET}
    params = {"query": query, "display": 10, "sort": "sim"}
    try:
        r = requests.get(url, headers=headers, params=params, timeout=7)
        r.raise_for_status()
        data = r.json()
        items = data.get("items", []) or []
        if not items:
            return None

        name_norm   = _normalize(place_name)
        region_norm = _normalize(region)
        best, best_score = None, -999
        for it in items:
            sc = _score_naver_item(it, name_norm, region_norm)
            if sc > best_score:
                best, best_score = it, sc

        if best and best_score >= 2:
            return best.get("link")
        return None
    except:
        return None

def get_og_image_from_kakao(place_url):
    """카카오 장소 상세 페이지의 og:image를 폴백으로 사용."""
    try:
        h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"}
        r = requests.get(place_url, headers=h, timeout=7)
        r.raise_for_status()
        html = r.text
        m = re.search(r'<meta[^>]+property=["\']og:image["\'][^>]+content=["\']([^"\']+)["\']', html, re.IGNORECASE)
        return m.group(1) if m else None
    except:
        return None

# ------------ 좌표 유틸 ------------
def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) 주소 검색
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) 키워드 그대로
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) 숫자 제거 후 키워드
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

# ------------ Selenium 평점 크롤링 ------------
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector)))
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

# ------------ 장소 검색 (이미지: 네이버 엄격 선택 → og:image 폴백) ------------
def get_places(category, address, size=15, sort="accuracy", radius=3000):
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_name = p.get("place_name")
        addr_q     = p.get("road_address_name") or p.get("address_name")
        place_url  = p.get("place_url")

        # 네이버 이미지(엄격) → 실패 시 og:image
        img = get_naver_image_strict(place_name, addr_q)
        if not img and place_url:
            img = get_og_image_from_kakao(place_url)

        rows.append({
            "이름": place_name,
            "카테고리": category,
            "주소": addr_q,
            "위도": float(p["y"]),
            "경도": float(p["x"]),
            "전화번호": p.get("phone"),
            "링크": place_url,
            "대표이미지": img
        })
    return pd.DataFrame(rows)

# ------------ 아이콘 & 카드 ------------
icon_urls = {
    "헬스장": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "공원": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "내과": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "종합병원": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "내분비내과": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty:
        return "<p style='color:#777'>표시할 결과가 없습니다.</p>"
    top5 = df.head(5)
    html = "<h3>📌 Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['대표이미지']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" if pd.notna(r.get("대표이미지")) and r.get("대표이미지") else ""
        rating_txt = "정보 없음"
        if "평점" in r and pd.notna(r["평점"]):
            try:
                rating_txt = f"{float(r['평점']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['이름']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">카테고리: {r['카테고리']}</p>
            <p style="margin:0;font-size:12px;color:#555;">⭐ 평점: {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['주소']}</p>
        </div>
        """
    return html

# ------------ 메인: 지도 생성 & 표시 ------------
def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000

    # 1) 평점 없이 장소 목록 먼저 수집
    dfs = []
    cats = ["헬스장", "공원", "내과", "종합병원", "내분비내과"] if category == "전체" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius)
        if not d.empty:
            dfs.append(d)

    # 지도의 기준 좌표
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780

    # 지도 생성
    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>내 위치</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

    # 2) 항상 평점 수집해서 표시 (정렬은 아래에서 선택적으로 적용)
    if not df.empty:
        driver = init_driver()
        try:
            ratings = get_kakao_ratings_bulk(driver, df["링크"].tolist())
            df["평점"] = df["링크"].map(ratings)
        finally:
            try: driver.quit()
            except: pass

        # 3) 정렬
        if sort == "rating" and "평점" in df.columns:
            df = df.sort_values("평점", ascending=False, na_position="last")

        # 마커 추가
        for row in df.itertuples():
            icon_url = icon_urls.get(row.카테고리)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.대표이미지}' width='150' style='border-radius:6px;object-fit:cover;'><br>" if getattr(row, "대표이미지", None) else ""
            rating_txt = "정보 없음"
            val = getattr(row, '평점', None)
            if pd.notna(val):
                try:
                    rating_txt = f"{float(val):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.이름}</b><br>"
                f"카테고리: {row.카테고리}<br>"
                f"주소: {row.주소}<br>"
                f"전화: {row.전화번호}<br>"
                f"⭐ 평점: {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.링크}' target='_blank'>카카오맵에서 보기</a>",
                max_width=320
            )
            folium.Marker([row.위도, row.경도], icon=icon, popup=popup).add_to(m)

    # 4) 저장 + 렌더
    os.makedirs("./_maps", exist_ok=True)
    map_path = f"./_maps/folium_map_{int(time.time())}.html"
    m.save(map_path)

    cards_html = top5_cards_html(df)
    full_html = f"""
    <div style="display:flex;gap:20px;align-items:flex-start;">
      <div style="flex:2;min-width:500px;">
        <iframe src="{map_path}" style="width:100%;height:600px;border:none;"></iframe>
      </div>
      <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
        {cards_html}
      </div>
    </div>
    """
    return full_html

# ------------ 위젯 UI ------------
address_input = Text(value="서울 종로구 세종대로 110", description="주소:", layout={'width':'60%'})
category_dd   = Dropdown(options=["전체","헬스장","공원","내과","종합병원","내분비내과"], value="전체", description="보기:")
sort_dd       = Dropdown(options=[("정확도순","accuracy"),("가까운순","distance"),("평점순","rating")], value="accuracy", description="정렬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="반경(km):")
btn_search    = Button(description="검색", button_style="success")
out           = Output()

def on_click(b):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
btn_search.on_click(on_click)
display(ui)
```

```python
import re
import folium
import pandas as pd
import requests, os, time
from typing import Optional
from ipywidgets import Text, Dropdown, IntSlider, Button, HBox, VBox, Output
from IPython.display import display, HTML, clear_output

# Selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# ====== 설정 ======
# 지도 렌더 방식: "inline"(권장) 또는 "iframe"(파일 저장 후 iframe)
RENDER_MODE = "inline"

# === API KEYS ===
KAKAO_API_KEY = "4f25ea2e2ce1c7a3a76f4ed96a3a44ff"
NAVER_CLIENT_ID = "ljQKyv2MAkTex1q2Vjeg"
NAVER_CLIENT_SECRET = "TjKu5MEvWH"

# -------------------------------
# (A) 네이버 이미지 API만 사용해 대표이미지 선택
# -------------------------------
def _normalize(s: str) -> str:
    if not isinstance(s, str):
        return ""
    return re.sub(r"[\s\-\_\|\(\)\[\]{}·•∙,.;:!?'\"<>@#%^&*+=/\\]", "", s.lower())

def _extract_region_token(addr_text: str) -> str:
    """주소에서 우선순위: 동 > 구 > 읍 > 면 한 개 토큰만 추출"""
    if not isinstance(addr_text, str):
        return ""
    toks = [t for t in addr_text.split() if t]
    for suf in ("동", "구", "읍", "면"):
        cand = [t for t in toks if t.endswith(suf)]
        if cand:
            return cand[0]
    return ""

def _is_good_ext(link: str) -> bool:
    base = link.lower().split("?")[0]
    return base.endswith((".jpg", ".jpeg", ".png", ".webp"))

def _score_naver_item(item, name_norm: str, region_norm: str) -> int:
    """제목/링크 토큰 매칭 + 해상도 가점, 음식/광고류는 즉시 탈락"""
    title = (item.get("title") or "")
    link  = (item.get("link") or "")
    w = int(item.get("sizewidth", 0) or 0)
    h = int(item.get("sizeheight", 0) or 0)

    if not link.startswith("http") or not _is_good_ext(link):
        return -999

    low_title = title.lower()
    low_link  = link.lower()
    # 음식/맛집/광고성 키워드 배제
    bad = ("치킨","맛집","식당","카페","메뉴","배달","분식","술집","맥주",
           "dessert","bread","recipe","food","mangoplate","siksin","coupon","banner","ad")
    if any(b in low_title or b in low_link for b in bad):
        return -999

    score = 0
    t_norm = _normalize(title)
    l_norm = _normalize(link)
    if name_norm and (name_norm in t_norm or name_norm in l_norm):
        score += 3
    if region_norm and (region_norm in t_norm or region_norm in l_norm):
        score += 2
    if w >= 400 and h >= 300:
        score += 1
    return score

def get_naver_image_only(place_name: str, addr_text: str) -> Optional[str]:
    """
    네이버 이미지 OpenAPI만 사용.
    쿼리: '{동/구} {상호명} 헬스/시설 관련 양의 키워드 -음식/맛집 부정 키워드'
    상위 10개 중 스코어가 가장 높은 링크를 대표이미지로 선택(스코어<2면 None).
    """
    region = _extract_region_token(addr_text)
    # 장소/시설 위주로 보이게 긍정 키워드, 음식/맛집 제거
    positive = "헬스장 체육관 실내 gym fitness 시설 내부"
    negative = "-치킨 -맛집 -식당 -카페 -메뉴 -배달 -분식 -술집 -맥주 -dessert -bread -recipe -food -mangoplate -siksin -coupon -banner -ad"
    base_query = f"{region} {place_name}".strip() if region else f"{place_name} {addr_text}".strip()
    query = f"{base_query} {positive} {negative}"

    url = "https://openapi.naver.com/v1/search/image"
    headers = {
        "X-Naver-Client-Id": NAVER_CLIENT_ID,
        "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
    }
    params = {"query": query, "display": 10, "sort": "sim"}

    try:
        r = requests.get(url, headers=headers, params=params, timeout=8)
        r.raise_for_status()
        items = r.json().get("items", []) or []
        if not items:
            return None
        name_norm   = _normalize(place_name)
        region_norm = _normalize(region)
        best, best_score = None, -999
        for it in items:
            sc = _score_naver_item(it, name_norm, region_norm)
            if sc > best_score:
                best, best_score = it, sc
        if best and best_score >= 2:
            return best.get("link")
        return None
    except:
        return None

# -----------------------------------
# (B) 카카오 좌표 변환
# -----------------------------------
def get_coords_from_address(address):
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    # 1) 주소 검색
    try:
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        r = requests.get(url, headers=headers, params={"query": address}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 2) 키워드
    try:
        url = "https://dapi.kakao.com/v2/local/search/keyword.json"
        r = requests.get(url, headers=headers, params={"query": address, "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    # 3) 숫자 제거 후 키워드
    try:
        short_address = "".join([c for c in address if not c.isdigit()])
        r = requests.get(url, headers=headers, params={"query": short_address.strip(), "size": 1}, timeout=7).json()
        if r.get("documents"):
            return float(r["documents"][0]["y"]), float(r["documents"][0]["x"])
    except:
        pass
    return None, None

# -------------------------------
# (C) Selenium 평점 크롤링
# -------------------------------
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1280,2000")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option("useAutomationExtension", False)
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36")
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def _extract_first_float(text):
    if not text:
        return None
    m = re.search(r'(\d+(?:\.\d+)?)', text)
    if m:
        try:
            return float(m.group(1))
        except:
            return None
    return None

def get_kakao_ratings_bulk(driver, place_urls):
    results = {}
    selectors = [
        "span.num_star", "span.txt_score", "em.num_rate", "strong.score_num",
        ".place_detail .evaluation .num", ".grade_star .num", "span.score_num",
        ".rating .num"
    ]
    for url in place_urls:
        rating = None
        try:
            driver.get(url)
            WebDriverWait(driver, 6).until(lambda d: d.execute_script("return document.readyState") == "complete")
            driver.execute_script("window.scrollTo(0, 400);")
            for selector in selectors:
                try:
                    elem = WebDriverWait(driver, 4).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                    )
                    r = _extract_first_float((elem.text or "").strip())
                    if r is not None:
                        rating = r
                        break
                except:
                    continue
        except:
            rating = None
        results[url] = rating
    return results

# ------------------------------------------
# (D) 카카오 API로 장소 검색 + 대표이미지(네이버만) / 평점
# ------------------------------------------
def get_places(category, address, size=15, sort="accuracy", radius=3000):
    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        return pd.DataFrame()

    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    params = {
        "query": category,
        "size": size,
        "x": lon, "y": lat, "radius": radius,
        "sort": "distance" if sort == "distance" else "accuracy"
    }
    try:
        res = requests.get(url, headers=headers, params=params, timeout=7).json()
    except:
        return pd.DataFrame()

    docs = res.get("documents", [])
    if not docs:
        return pd.DataFrame()

    rows = []
    for p in docs:
        place_name = p.get("place_name")
        addr_q     = p.get("road_address_name") or p.get("address_name")
        place_url  = p.get("place_url")

        # ✅ 대표이미지: 네이버 이미지 API만 사용
        img = get_naver_image_only(place_name, addr_q)

        rows.append({
            "이름": place_name,
            "카테고리": category,
            "주소": addr_q,
            "위도": float(p["y"]),
            "경도": float(p["x"]),
            "전화번호": p.get("phone"),
            "링크": place_url,
            "대표이미지": img
        })
    return pd.DataFrame(rows)

# -------------------
# (E) 아이콘 & 카드
# -------------------
icon_urls = {
    "헬스장": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/weightlifter.png",
    "공원": "https://github.com/seonghuyn/icon_list/blob/master/park%20(1).png?raw=true",
    "내과": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/stethoscope1.png",
    "종합병원": "https://raw.githubusercontent.com/seonghuyn/icon_list/master/hospital.png",
    "내분비내과": "https://img.icons8.com/ios/50/pills.png"
}

def top5_cards_html(df):
    if df is None or df.empty:
        return "<p style='color:#777'>표시할 결과가 없습니다.</p>"
    top5 = df.head(5)
    html = "<h3>📌 Top 5</h3>"
    for _, r in top5.iterrows():
        img_tag = f"<img src='{r['대표이미지']}' style='width:100%;height:120px;object-fit:cover;border-radius:8px;'>" \
                  if pd.notna(r.get("대표이미지")) and r.get("대표이미지") else ""
        rating_txt = "정보 없음"
        if "평점" in r and pd.notna(r["평점"]):
            try:
                rating_txt = f"{float(r['평점']):.1f}"
            except:
                pass
        html += f"""
        <div style="width:220px;border:1px solid #ddd;border-radius:10px;padding:10px;margin-bottom:10px;box-shadow:2px 2px 5px #ccc;">
            {img_tag}
            <h4 style="margin:8px 0 4px 0;font-size:14px;">{r['이름']}</h4>
            <p style="margin:0;font-size:12px;color:#555;">카테고리: {r['카테고리']}</p>
            <p style="margin:0;font-size:12px;color:#555;">⭐ 평점: {rating_txt}</p>
            <p style="margin:0;font-size:11px;color:#777;">{r['주소']}</p>
        </div>
        """
    return html

# --------------------------------
# (F) 지도 생성 & 표시 (자동 갱신/카테고리 확실 필터)
# --------------------------------
def build_and_show(address, category, sort, radius_km):
    radius = int(radius_km) * 1000

    # 데이터 수집 (이미지 포함)
    dfs = []
    cats = ["헬스장", "공원", "내과", "종합병원", "내분비내과"] if category == "전체" else [category]
    for cat in cats:
        d = get_places(cat, address, size=15, sort=sort, radius=radius)
        if not d.empty:
            dfs.append(d)

    lat, lon = get_coords_from_address(address)
    if not lat or not lon:
        lat, lon = 37.5665, 126.9780  # fallback

    m = folium.Map(location=[lat, lon], zoom_start=14)
    folium.Marker([lat, lon], popup=f"<b>내 위치</b><br>{address}",
                  icon=folium.Icon(color="red", icon="home")).add_to(m)

    df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

    # ✅ 최종 안전장치: 선택 카테고리만 남기기
    if not df.empty and category != "전체":
        df = df[df["카테고리"] == category].copy()

    # 평점은 항상 수집해서 표시, '평점순'일 때만 정렬
    if not df.empty:
        driver = init_driver()
        try:
            ratings = get_kakao_ratings_bulk(driver, df["링크"].tolist())
            df["평점"] = df["링크"].map(ratings)
        finally:
            try: driver.quit()
            except: pass

        if sort == "rating" and "평점" in df.columns:
            df = df.sort_values("평점", ascending=False, na_position="last")

        for row in df.itertuples():
            icon_url = icon_urls.get(row.카테고리)
            icon = folium.CustomIcon(icon_url, icon_size=(35, 35)) if icon_url else None
            img_html = f"<img src='{row.대표이미지}' width='150' style='border-radius:6px;object-fit:cover;'><br>" \
                       if getattr(row, "대표이미지", None) else ""
            rating_txt = "정보 없음"
            val = getattr(row, '평점', None)
            if pd.notna(val):
                try:
                    rating_txt = f"{float(val):.1f}"
                except:
                    pass
            popup = folium.Popup(
                f"<b>{row.이름}</b><br>"
                f"카테고리: {row.카테고리}<br>"
                f"주소: {row.주소}<br>"
                f"전화: {row.전화번호 or '-'}<br>"
                f"⭐ 평점: {rating_txt}<br>"
                f"{img_html}"
                f"<a href='{row.링크}' target='_blank' rel='noopener noreferrer'>카카오맵에서 보기</a>",
                max_width=320, parse_html=True
            )
            folium.Marker([row.위도, row.경도], icon=icon, popup=popup).add_to(m)

    # 렌더
    cards_html = top5_cards_html(df)
    if RENDER_MODE == "iframe":
        os.makedirs("./_maps", exist_ok=True)
        map_path = f"./_maps/folium_map_{int(time.time())}.html"
        m.save(map_path)
        full_html = f"""
        <div style="display:flex;gap:20px;align-items:flex-start;">
          <div style="flex:2;min-width:500px;">
            <iframe src="{map_path}"
                    style="width:100%;height:600px;border:none;"
                    sandbox="allow-scripts allow-popups allow-popups-to-escape-sandbox allow-top-navigation-by-user-activation">
            </iframe>
          </div>
          <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
            {cards_html}
          </div>
        </div>
        """
        return full_html
    else:
        map_html = m._repr_html_()
        full_html = f"""
        <div style="display:flex;gap:20px;align-items:flex-start;">
          <div style="flex:2;min-width:500px;">{map_html}</div>
          <div style="flex:1;max-width:280px;overflow:auto;height:600px;">
            {cards_html}
          </div>
        </div>
        """
        return full_html

# ------------ 위젯 UI (자동 갱신) ------------
address_input = Text(value="대구시 달서구 월성동 월성e편한세상", description="주소:", layout={'width':'60%'})
category_dd   = Dropdown(options=["전체","헬스장","공원","내과","종합병원","내분비내과"], value="전체", description="보기:")
sort_dd       = Dropdown(options=[("정확도순","accuracy"),("가까운순","distance"),("평점순","rating")], value="accuracy", description="정렬:")
radius_slider = IntSlider(value=3, min=1, max=20, step=1, description="반경(km):")
btn_search    = Button(description="검색", button_style="success")
out           = Output()

def run_search(*args):
    with out:
        clear_output()
        html = build_and_show(address_input.value, category_dd.value, sort_dd.value, radius_slider.value)
        display(HTML(html))

# 버튼 클릭
btn_search.on_click(lambda b: run_search())

# ✅ 위젯 변경 즉시 자동 업데이트
category_dd.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
sort_dd.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
radius_slider.observe(lambda change: run_search() if change['name']=='value' else None, names='value')
address_input.observe(lambda change: run_search() if change['name']=='value' else None, names='value')

# UI 렌더 + 최초 1회 실행
ui = VBox([HBox([address_input, btn_search]), category_dd, sort_dd, radius_slider, out])
display(ui)
run_search()
```
